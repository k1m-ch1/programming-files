```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

# Discrete Random Variables

## Fair Dice
- A fair Dice has 6 outcomes with a uniform distribution

|x|1|2|3|4|5|6|
|---|---|---|---|---|---|---|
|P(X=x)|$\frac{1}{6}$|$\frac{1}{6}$|$\frac{1}{6}$|$\frac{1}{6}$|$\frac{1}{6}$|$\frac{1}{6}$|

```python
xk = np.arange(1, 7)
pk = np.full(len(xk), 1/len(xk))

dice = st.rv_discrete(name="dice", values=(xk, pk))
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
unique, counts = np.unique(dice.rvs(size=100000), return_counts=True)
ax[0].bar(unique, counts, label="true dice throw")
ax[1].bar(unique, np.sum(counts)*dice.pmf(unique), label="expected dice throw", color="orange")
fig.legend()
```
