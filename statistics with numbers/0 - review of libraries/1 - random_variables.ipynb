# Discrete Random Variables

## Fair Dice
- A fair Dice has 6 outcomes with a uniform distribution

|x|1|2|3|4|5|6|
|---|---|---|---|---|---|---|
|P(X=x)|$\frac{1}{6}$|$\frac{1}{6}$|$\frac{1}{6}$|$\frac{1}{6}$|$\frac{1}{6}$|$\frac{1}{6}$|

```python
class dice_gen(st.rv_discrete):
    def _pmf(self, n:int):
        if n in np.arange(1, 7, 1):
            return 1/6
        else:
            return 0
            
dice = dice_gen(name="dice")
dice._pmf()
```
