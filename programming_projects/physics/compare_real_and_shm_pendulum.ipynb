```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
domain_min = 0
domain_max = 180
theta_zero = np.pi
omega = np.pi/4


t = np.linspace(0, domain_max, 10000)
y1 = theta_zero * np.cos(omega * t)

def generate_plot(initial, domain_min, domain_max, dt):
    list_of_t = [0]
    list_of_theta = [initial]
    list_of_theta_dot = [0]
    theta_dot = lambda prev_theta, prev_theta_dot: prev_theta_dot - (omega**2)*np.sin(prev_theta)*dt
    theta = lambda prev_theta, theta_dot: prev_theta + theta_dot*dt
    for _ in range(round((domain_max - domain_min)/dt)):
        list_of_t.append(list_of_t[-1]+dt)
        list_of_theta_dot.append(theta_dot(list_of_theta[-1], list_of_theta_dot[-1]))
        list_of_theta.append(theta(list_of_theta[-1], list_of_theta_dot[-1]))

    return list_of_t, list_of_theta

list_of_t, list_of_theta = generate_plot(theta_zero, domain_min, domain_max, 0.1)
# Create subplots
fig, axs = plt.subplots(1, 1, figsize=(8, 6))  # 3 rows, 1 column

print(max(list_of_theta))

# Plot data on each subplot
axs.plot(t, y1, color='r')
axs.set_title('SHM vs Real')
axs.plot(list_of_t, list_of_theta, color='b')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
